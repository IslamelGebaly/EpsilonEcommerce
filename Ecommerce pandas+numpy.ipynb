Ask yourself 15 analystical questions that you belive will help the managers in decision making

```python
import pandas as pd
import numpy as np
```

```python
ecommerce_df = pd.read_csv("ecommerce.csv")
ecommerce_df.head()
```

```python
ecommerce_df.info()
```

```python
ecommerce_df.describe()
```

```python
ecommerce_df.rename(columns = {"Quantit" : "Quantity"}, inplace=True)
ecommerce_df.columns = ecommerce_df.columns.str.strip()
```

```python
ecommerce_df["Year"] = pd.to_datetime(ecommerce_df["Year-Month"]).dt.year
ecommerce_df["Month"] = pd.to_datetime(ecommerce_df["Year-Month"]).dt.month
ecommerce_df["InvoiceTime"] = pd.to_datetime(ecommerce_df["InvoiceTime"].str.strip(), format = "%H:%M")
ecommerce_df["InvoiceHour"] = ecommerce_df["InvoiceTime"].dt.hour
```

```python
ecommerce_df["Country"] = ecommerce_df["Country"].str.strip()
ecommerce_df["Country"].unique()
```

```python
ecommerce_df.head()
```

## Q1. What is the Average Order Value per Country?

```python
ecommerce_df.groupby("Country").agg({
    "OrderValue" : "mean"
}).sort_values("OrderValue", ascending=False)
```

## Q2. Which Major Category has the highest Average Order Value?

```python
ecommerce_df.groupby("Major Category").agg({
    "OrderValue" : "mean"
}).sort_values("OrderValue", ascending=False).head(1)
```

## Q3. What is the highest Selling Minor Category?

```python
ecommerce_df.groupby("Minor Category").agg({
    "OrderValue" : "sum"
}).sort_values("OrderValue",ascending=False).head(1)
```

## Q4. What is the average order value per month for each country

```python
ecommerce_df.groupby(["Country", "Month"]).agg({
    "OrderValue" : "mean"
}).sort_values(["Country", "Month"], ascending=True)
```

## Q5. What is the total revenue for each year?

```python
ecommerce_df.groupby("Year").agg({
    "OrderValue" : "sum"
})
```

## Q6. Pivot Table showing the total revenue for per month for each year

```python
ecommerce_df.pivot_table(values= "OrderValue",index= "Year", columns="Month")
```

## Q7. What is the Largest Purchase made in 2010?

```python
ecommerce_df.loc[ecommerce_df["Year"] == 2010].sort_values("OrderValue", ascending = False).head(1)
```

## Q8. What is the Minor Category has the lowest revenue in the united kingdom?

```python
ecommerce_df.loc[ecommerce_df["Country"] == "United Kingdom"].groupby("Minor Category").agg({
    "OrderValue" : "sum"
}).sort_values("OrderValue", ascending=True).head(1)
```

## Q9. What are the top 10 items by quantity ordered?

```python
ecommerce_df.groupby("Description").agg({
    "Quantity" : "sum"
}).sort_values("Quantity", ascending= False).head(10)
```

## Q10. Pivot table that shows the number of order for each major category per month

```python
ecommerce_df.pivot_table(index="Major Category", values = "InvoiceNo", columns= "Month", aggfunc="count")
```

## Q11. What is the number of orders for each Country?

```python
ecommerce_df.groupby("Country").agg({
    "InvoiceNo" : "count"
})
```

## Q12. What is the number of orders per Major Category for each Month?

```python
ecommerce_df.groupby(["Major Category", "Month"]).agg({
    "InvoiceNo" : "count"
})
```

## Q13. At what hours do most orders happen?

```python
ecommerce_df.groupby("InvoiceHour").agg({
    "InvoiceNo" : "count"
}).sort_values("InvoiceNo", ascending=False).head(3)
```

## Q14. Pivot Table that shows the average unit price for each country per category

```python
ecommerce_df.pivot_table(index="Country", columns="Major Category", values= "UnitPrice", aggfunc="mean")
```

## Q15. What is the correlation between order value, unit price, and Quantity

```python
ecommerce_df[["OrderValue", "UnitPrice", "Quantity"]].corr()
```
